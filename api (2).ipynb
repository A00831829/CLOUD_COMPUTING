```python
import pandas as pd
import json
import requests
import numpy as np

# Cargar datos de ejemplo
import kagglehub
path = kagglehub.dataset_download("fedesoriano/company-bankruptcy-prediction")
datos = pd.read_csv(path + "/data.csv")
datos = datos.drop(["Bankrupt?"], axis=1).head(20)

# Mostrar vista previa
print("\nDatos que se enviarán al modelo:")
print(datos.head())

# Preparar datos para la API
data_dict = datos.to_dict(orient='list')
data_json = json.dumps({"data": [data_dict]})

# Cargar URI del modelo
with open("uri.json", "r") as suri:
    scoring_uri = json.load(suri)["URI"][0]
    print(f"URI cargada: {scoring_uri}")

# Cargar clave (token) desde archivo token.json
with open("token.json", "r") as stok:
    token = json.load(stok)["token"]  # Asegúrate que el archivo tenga {"token": "TU_CLAVE"}

# Realizar la solicitud a la API con autenticación
headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {token}"
}

print(f"Enviando solicitud a {scoring_uri}...")
response = requests.post(scoring_uri, data=data_json, headers=headers)

if response.status_code == 200:
    # Procesar la respuesta
    result = json.loads(response.json())
    print("Respuesta recibida correctamente.")

    # Añadir resultados al dataframe
    datos["Bancarrota_Predicha"] = result

    # Mostrar resultados
    print("\nResultados de la predicción:")
    print(datos[["Bancarrota_Predicha"] + list(datos.columns[:5])])
else:
    print(f"Error en la solicitud: {response.status_code}")
    print(f"Respuesta: {response.text}")
```
